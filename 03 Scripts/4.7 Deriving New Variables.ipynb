```python
df = ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
## conditions that apply in the price filter

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range product'
  else: return np.nan
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python

df['price_range'].value_counts(dropna = False)
```

```python
## identifying the most expensive product

df['prices'].max()
```

## If-Statements with the loc() Function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Create new if-statement for-loop for new requirements
result2 = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    result2.append("Busiest days")
  elif value == 4 or value == 3:
    result2.append("Least busy days")
  else:
    result2.append("Regularly busy")
```

## Exercise 4.7.4

```python
# Check frequency of values in the "order_hour_of_day" column 
ords_prods_merged['hour_of_day'].value_counts()
```

```python
## for-loop if statement using the above output
hour_of_day_new = []

for value in ords_prods_merged["hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hour_of_day_new.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hour_of_day_new.append("Fewest orders")
  else:
    hour_of_day_new.append("Average orders")
```

```python
ords_prods_merged['busiest_period_of_day'] = hour_of_day_new
```
