#### 1. Importing Datasets

```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
path = r'C:\Users\matth\05-2023 Instacart Basket Analysis'
```

```python
df_ords_comb = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.csv'), index_col = False)
```

```python
df_ords_comb.shape
```

```python
df_prod_clean = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_clean.csv'), index_col = False)
```

#### 2. Checking the datsets

```python
## checking the headers

df_ords_comb.head()
```

```python
df_ords_comb = df_ords_comb.drop(columns = ['Unnamed: 0'])
```

```python
df_prod_clean = df_prod_clean.drop(columns = ['Unnamed: 0'])
```

```python
df_prod_clean.head()
```

```python
df_ords_comb.head()
```

```python
## identifying shape of dataframe

df_ords_comb.shape
```

```python
df_prod_clean.shape
```

#### 3. Executing the merging function

```python
## merging the products and the orders combined data

df_merged_latest = df_ords_comb.merge(df_prod_clean, on = 'product_id')
```

```python
df_merged_latest
```

```python
## checking the results using merge flag

df_merged_latest['_merge'].value_counts()
```

#### 4. Exporting the dataset

```python
## exporting as .csv

df_merged_latest.to_csv(os.path.join(path, '02 Data','Prepared Data', 'orders_products_merged.csv'))
```

```python
## exporting as pkl

df_merged_latest.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_merged.pkl'))
```
